```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:2], df.iloc[:,-1], test_size=0.2)
```

```python
X_train.head()
```

```python
# Ordninal encoding r use for input/X_train dataset
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])
```

```python
# fit()  Learn something from data (like memorize info)
#    But it doesnt transform the data yet.
#    It just stores this mapping internally.

# transform()  Use that learned info to change data
     # It uses the stored mapping to actually convert

# fit_transform()  Do both together (learn + change)
     # This is shortcut version  it does:
        #   fit()  learn mapping
        #   transform()  apply mapping
```

```python
oe.fit(X_train)
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
